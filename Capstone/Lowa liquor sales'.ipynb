```python
df1['date'] = pd.to_datetime(df1['date'])
df1.info()
```

```python
df1['Year'] = df1['date'].dt.year.astype('object')
     

# Per Quarter of the Year.
df1['Quarter'] = df1['date'].dt.quarter.astype('object')
     

# Month of the Year.
df1['Month'] = df1['date'].dt.month.astype('object')
     

# Week of the Year.
df1['Week'] = df1['date'].dt.isocalendar().week.astype('object')
     

# Drop 'Date' column as is not needed anymore.
#df = df.drop(columns='Date')
```

```python
df1.info()
```

```python
df1.drop(columns = ['invoice_and_item_number',
                    'category_name', 'vendor_name',
                    'address', 'zip_code',
                   'store_location', 'category', 'county_number',
                   'vendor_number', 'item_number', 'item_description'],
        inplace = True)
```

```python
df1.columns
```

```python
num_colums=df1.select_dtypes(include = 'number')
```

```python
num_colums.columns
```

```python
fig,axs = plt.subplots(3,3,figsize =(15,15))
for i, subplots in zip(num_colums,axs.flatten()):
    sns.boxplot(df1[i],ax=subplots)
    plt.title('df1[i]',fontsize = 15)
plt.show()
```

```python
for i in num_colums:
    q1,q3 = np.quantile(df1[i],[0.25,0.75])
    iqr = q3 - q1
    ub = q3 + (1.5 * iqr)
    lb = q1 - (1.5 * iqr)
    df1[i] = np.where(df1[i] > ub, ub, df1[i])
    df1[i] = np.where(df1[i] < lb, lb, df1[i])
    
```

```python
fig,axs = plt.subplots(3,3,figsize =(15,15))
for i, subplots in zip(num_colums,axs.flatten()):
    sns.boxplot(df1[i],ax=subplots)
    plt.title('df1[i]',fontsize = 15)
plt.show()
```

```python
fig,axs = plt.subplots(3,3,figsize =(15,15))
for i, subplots in zip(num_colums,axs.flatten()):
    sns.violinplot(df1[i],ax=subplots)
    plt.title('df1[i]',fontsize = 15)
plt.show()
```

```python
df1.groupby('county').agg(value_count=('county','count')).sort_values(by=['county','value_count'], ascending=[True,False])
```
